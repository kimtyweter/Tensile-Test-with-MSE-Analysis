### 2mm All

```python
data2 = [df_RD_2mm, df_DD_2mm, df_TD_2mm]
# data = [df_RD_1mm, df_DD_1mm, df_TD_1mm, df_RD_2mm, df_DD_2mm, df_TD_2mm]
n_value = [0.321, 0.322, 0.307]
K_value = [492.9, 482.8, 473.92]
yield_value = [115.527, 111.803, 113.587]
e_y_value = [0.000109, 0.000106, 0.000118]
```

```python
ind = 0
for i in data2:
    # change column name to plot
    list_name = i.columns.tolist()
    name_file = list_name[1]
    list_name[1] = 'Stress'
    i.columns = list_name
    
    # 
    
    i['log_Stress'] = np.log10(i['Stress'])
    i['log_Strain'] = np.log10(i['Engineering Strain'])
    i['log_M1_Prediction_Stress'] = np.log10(K_value[ind]) + n_value[ind] * np.log10(i['Engineering Strain'])

    i_by = i[i['Engineering Strain'] <= e_y_value[ind]]
    i_by = i_by[-i_by['log_Stress'].isna()]
    i_by['ln_delta'] = np.log(abs(i['log_M1_Prediction_Stress'] - i['log_Stress'] ))
    
    
    x = np.array(i_by['log_Strain'].tolist())
    y = np.array(i_by['ln_delta'].tolist())
    K2, n2 = np.polyfit(x, y, 1)
    
    # Model 2:
    i['M2_Prediction_Stress'] = K_value[ind] * (i['Engineering Strain']) ** n_value[ind] + np.exp(K2 + n2 * i['Engineering Strain'])
    ind += 1


# Plot
# Prediction vs experiment
ax = sns.lineplot(df_RD_2mm["Engineering Strain"], df_RD_2mm["Stress"], color = 'red')
ax = sns.lineplot(df_DD_2mm["Engineering Strain"], df_DD_2mm["Stress"], color = 'gray')
ax = sns.lineplot(df_TD_2mm["Engineering Strain"], df_TD_2mm["Stress"], color = 'steelblue')
ax = sns.lineplot(df_RD_2mm["Engineering Strain"], df_RD_2mm["M2_Prediction_Stress"], color = 'green')
ax = sns.lineplot(df_DD_2mm["Engineering Strain"], df_DD_2mm["M2_Prediction_Stress"], color = 'orange')
ax = sns.lineplot(df_TD_2mm["Engineering Strain"], df_TD_2mm["M2_Prediction_Stress"], color = 'purple')
import matplotlib.patches as mpatches
a = mpatches.Patch(color='red', label='Experiment-RD')
b = mpatches.Patch(color='gray', label='Experiment-DD')
c = mpatches.Patch(color='steelblue', label='Experiment-TD')
d = mpatches.Patch(color='green', label='Prediction-RD')
e = mpatches.Patch(color='orange', label='Prediction-DD')
f = mpatches.Patch(color='purple', label='Prediction-TD')

plt.legend(handles=[d,e,f,a,c,b],fontsize = 10, framealpha=0,facecolor='white', edgecolor = 'black')
plt.ylabel('Stress', fontsize = 18)
plt.xlabel('Strain', fontsize = 18)
plt.title('2mm', fontsize = 24)
plt.savefig('M2_2mm.jpg', dpi=300, transparent=False, bbox_inches="tight")
```

```python
# compute residual
df_RD_2mm['M2_Residual'] = abs(df_RD_2mm['M2_Prediction_Stress'] - df_RD_2mm['Stress'])
df_DD_2mm['M2_Residual'] = abs(df_DD_2mm['M2_Prediction_Stress'] - df_DD_2mm['Stress'])
df_TD_2mm['M2_Residual'] = abs(df_TD_2mm['M2_Prediction_Stress'] - df_TD_2mm['Stress'])
```

```python
res_m1 = []
res_m2 = []
res_m3 = []
res_m4 = []
res_m5 = []
res_m6 = []
from sklearn.metrics import mean_squared_error
for i in data1:
    res_m1.append(mean_squared_error(i['Stress'], i['M1_Prediction_Stress']))
    res_m2.append(mean_squared_error(i['Stress'], i['M2_Prediction_Stress']))
    
for i in data2:
    res_m1.append(mean_squared_error(i['Stress'], i['M1_Prediction_Stress']))
    res_m2.append(mean_squared_error(i['Stress'], i['M2_Prediction_Stress']))

for i in data3:
    res_m3.append(mean_squared_error(i['Stress'], i['Swift hardening model']))
    res_m4.append(mean_squared_error(i['Stress'], i['Voce constitutive model']))
    res_m5.append(mean_squared_error(i['Stress'], i['Modified Voce hardening model']))
    res_m6.append(mean_squared_error(i['Stress'], i['Kim-Tuan hardening model']))
    
```
